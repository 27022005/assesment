```python
# 4. Are there any missing or corrupted entries in Ship Date, Order Date, or numeric columns?
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

df[['Order Date', 'Ship Date']].isnull().sum()
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[numeric_cols].isnull().sum()
```

```python
# 5. Convert Order Date and Ship Date to datetime.
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

df.dtypes[['Order Date', 'Ship Date']]
```

```python
# 8. What are the unique values in Customer Type and Order Priority?
unique_order_priority = df["Order Priority"].unique()
unique_customer_type = df["Customer Type"].unique()

print("unique order priority is", unique_order_priority)
print("unique customet type is",unique_customer_type)
```

```python
# 11. What’s the range of order quantities and prices?
min_of_order_qty = df["Order Quantity"].min()
max_of_order_qty = df["Total"].max()
rage_of_order_qty = (max_of_order_qty - min_of_order_qty)
print("rage of order quantity is:-",rage_of_order_qty)

min_of_price = df["Total"].min()
max_of_price = df["Total"].max()
rage_of_price = max_of_price - min_of_price
print("rage of price is:-",rage_of_price)

```

```python
# 12. Create a new column for shipping duration.

df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

df["Shipping Duration"] = (df["Ship Date"] - df["Order Date"]).dt.days
df
```

```python
# 13. Are there any orders with zero or negative total or quantity?
zero_qty = df[df["Order Quantity"] == 0]
negative_qty = df[df["Order Quantity"] < 0]
print("Orders with ZERO quantity:\n", zero_qty)
print("\nOrders with NEGATIVE quantity:\n", negative_qty)

df["Total"] = pd.to_numeric(df["Total"], errors='coerce')

zero_total = df[df["Total"] == 0]
negative_total = df[df["Total"] < 0]
print("\nOrders with ZERO total price:\n", zero_total)
print("\nOrders with NEGATIVE total price:\n", negative_total)
```

```python
# 14. Are all discount percentages matching discount dollar amounts?
cols = ['Sub Total', 'Discount %', 'Discount $']

for col in cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r'[$,]', '', regex=True)
        .replace('nan', None)
        .astype(float)
    )

df['Expected Discount $'] = (df['Discount %'] / 100) * df['Sub Total']

valid_discount_rows = df.dropna(subset=['Sub Total', 'Discount %', 'Discount $'])

valid_discount_rows['Discount Match'] = (
    valid_discount_rows['Expected Discount $'].astype(float).round(2) ==
    valid_discount_rows['Discount $'].astype(float).round(2))

print("Do all valid discount percentages match the discount dollar amounts?",
      valid_discount_rows['Discount Match'].all())

valid_discount_rows[valid_discount_rows['Discount Match'] == False][
    ['Sub Total', 'Discount %', 'Discount $', 'Expected Discount $']]


```

```python
# 15. Check for mismatches in total calculation.
cols = ['Order Total', 'Shipping Cost', 'Total']

for col in cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r'[$,]', '', regex=True)
        .replace('nan', None)
        .astype(float)
    )
df['Expected Total'] = df['Order Total'] + df['Shipping Cost']
valid_total_rows = df.dropna(subset=['Order Total', 'Shipping Cost', 'Total'])

valid_total_rows['Total Match'] = (
    valid_total_rows['Expected Total'].round(2) ==
    valid_total_rows['Total'].round(2)
)
print("Do all valid total values match their calculated totals?",
      valid_total_rows['Total Match'].all())

valid_total_rows[valid_total_rows['Total Match'] == False][
    ['Order Total', 'Shipping Cost', 'Total', 'Expected Total']
]
```

```python
# 25. What is the most profitable product overall?
df["Total"] = pd.to_numeric(df["Total"], errors="coerce")
df["Cost Price"] = pd.to_numeric(df["Cost Price"], errors="coerce")
df["Order Quantity"] = pd.to_numeric(df["Order Quantity"], errors="coerce")

df["Profit"] = df["Total"] - (df["Cost Price"] * df["Order Quantity"])

most_profitable = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False)

print("Most Profitable Product:")
print(most_profitable.head(1))
```

```python
# 27. Do longer shipping times impact profit margins?
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

df['Shipping Time (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Profit Margin'] = (
    df['Profit Margin']
    .astype(str)
    .str.replace(r'[$,]', '', regex=True)
    .replace('nan', None)
    .astype(float)
)
correlation = df[['Shipping Time (Days)', 'Profit Margin']].corr()

print("Correlation between Shipping Time and Profit Margin:")
print(correlation)
df['Shipping Speed'] = pd.cut(
    df['Shipping Time (Days)'],
    bins=[0, 3, 7, 100],
    labels=['Fast', 'Medium', 'Slow']
)
avg_profit_by_speed = df.groupby('Shipping Speed')['Profit Margin'].mean()

print("\nAverage Profit Margin by Shipping Speed:")
print(avg_profit_by_speed)
```
