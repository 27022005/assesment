```python
# Module 8: NumPy, Pandas, Matplotlib Libraries  
# ASSESMENT:- 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\DELL\Documents\pallavi's file\python assesments\historical_automobile_sales.csv")
df
```

```python
# Q1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year.
p = df.groupby("Year")["Automobile_Sales"].sum()
ph = p.reset_index()
plt.figure(figsize=(10,5))
plt.plot(ph["Year"],ph["Automobile_Sales"],marker=".",color="red",linestyle="-")
plt.title("Yearly Automobile Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
# Q2: Plot different lines for categories of vehicle type and analyze the trend to answer the question:
# Is there a noticeable difference in sales trends between different vehicle types during recession periods?
df1 = df.groupby(["Vehicle_Type","Year"])["Automobile_Sales"].mean().reset_index()
plt.figure(figsize=(10,6))
for i in df1["Vehicle_Type"].unique():
    temp = df1[df1["Vehicle_Type"]==i]
    plt.plot(temp["Year"],temp["Automobile_Sales"],label=i)
plt.legend()
plt.title("Sales Trebd by vehical")
plt.xlabel("Year")
plt.ylabel("Vehicle_Type")
plt.show()
```

```python
# Q3: Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.
plt.Figure(figsize=(10,6))
sns.lineplot(data=df,x="Year",y="Automobile_Sales",hue="Vehicle_Type",style="Recession")
plt.title("Recession vs Non-Recession Sales Trend")
plt.show()
```

```python
# Q4: Compare the sales of different vehicle types during a recession and a non-recession period.
# Hint: To visualize sales of different vehicles during recession and non-recession periods, use a bar chart.
# Group by Recession and Vehicle_Type for average Automobile_Sales and then plot it using sns.barplot().
df2 = df.groupby(["Recession","Vehicle_Type"])["Automobile_Sales"].mean().reset_index()
plt.figure(figsize=(10,7))
sns.barplot(data=df,x="Vehicle_Type",y="Automobile_Sales",hue="Recession")
plt.title("Avg Sales of Vehicle Types: Recession vs Non-Recession")
plt.show()
```
